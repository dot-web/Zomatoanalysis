```python
df_country=pd.read_excel('/content/Country-Code.xlsx')
df_country.head()
```

```python
final_df=pd.merge(df,df_country)
```

```python
final_df.head(2)
```

```python
final_df.dtypes
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

```python
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.f%%')
```

Observation:
1. India has the most number of zomato users. It covers 94% of the total users in the world.
2. United states and United Kingdom follow the most number of users after India resperctively.


```python
Ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
Ratings
```

**Observations**
1. When the rating is 0.1-2.4 the color is Red and the food is poor.
2. When the rating is 2.4-3.4 the color is Orange and the food is Average.
3. When the rating is 4.0-3.9 the color is Yellow and the food is Good.
4. When the rating is 4.0-4.4 the color is Green and the food is very good.
5. When the rating is 4.4-4.9 the color is Dark green and the food is excellenet.
6. Most of the users did not give any rating

```python

plt.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y='Rating Count',data=Ratings)
```

```python
sns.barplot(x='Aggregate rating',y='Rating Count',hue='Rating color',data=Ratings,palette=['white','red','orange','yellow','green','green']);
```

**Observation**
1. No rating has the highest value
2. Most rating fall between 2.6-3.4

```python
sns.countplot(x='Rating color',data=Ratings,palette=['white','red','orange','yellow','green','green'])
```
